create filtered dataset: use the labels in 'selected_labels.csv'

```python
import os
import shutil
from tqdm import tqdm 
import pandas as pd
```

```python
sel_labels = pd.read_csv("selected_labels.csv", index_col=0)
sel_labels.head()
```

```python
source_folder = "Dataset"

destination_folder = "Filtered_Dataset"
os.makedirs(destination_folder, exist_ok=True)

filenames_to_keep = [f"{str(idx)}.jpg" for idx in sel_labels.index]

# Loop through and copy only the desired files
for filename in tqdm(filenames_to_keep):
    source_path = os.path.join(source_folder, filename)
    dest_path = os.path.join(destination_folder, filename)
    if os.path.exists(source_path):
        shutil.copy(source_path, dest_path)
    else:
        print(f"File not found: {filename}")
```
